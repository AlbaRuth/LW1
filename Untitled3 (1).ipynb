```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram, cophenet, fcluster
from scipy.spatial.distance import pdist
from scipy.spatial.distance import seuclidean, cityblock, chebyshev, minkowski, euclidean, mahalanobis  # Импортируем стандартизированную евклидову метрику

with open('/content/2023_2.txt', 'r') as file:
    lines = file.readlines()

data = [line.strip().split() for line in lines]
df = pd.DataFrame(data, columns=['X', 'Y'])
df[['X', 'Y']] = df[['X', 'Y']].apply(pd.to_numeric)

# Построение графика рассеяния
plt.scatter(df['X'], df['Y'], c='c')

# Здесь настраивается сетка на графике, чтобы делать его более читаемым.
plt.rc('grid', linestyle=':', linewidth=0.5)
plt.grid(True)
# Добавление подписей осей
plt.xlabel('X')
plt.ylabel('Y')

# Добавление заголовка
plt.title('График рассеяния для данных X и Y')

# Отображение графика
plt.show()

# Выполняется иерархическая кластеризация методом дальнего соседа с использованием стандартизированной евклидовой метрики расстояния.
data = df[['X', 'Y']].to_numpy()

# Создается матрица связей между точками.
linkage_matrix = linkage(data, method='single', metric='euclidean') #method(complete/average/centroid)- метод связывания | metric(seuclidean/cityblock) - метрика

dendrogram(linkage_matrix, labels=range(len(data)), orientation='top')
plt.xlabel('Индексы точек')
plt.ylabel('Расстояние')
plt.title('Дендрограмма иерархической кластеризации')
plt.show()

# Рассчитываем матрицу попарных расстояний между точками
distance_matrix = pdist(data, metric='euclidean')  # Устанавливаем евклидову метрику расстояния

# Рассчитываем кофенетический коэффициент
c, coph_dists = cophenet(linkage_matrix, distance_matrix)
print(f"Кофенетический коэффициент: {c}")

# Определяем кластеры на основе порога
threshold = 3  # Измените этот порог при необходимости
clusters = fcluster(linkage_matrix, threshold, criterion='distance')

# Вычисляем центры кластеров
cluster_centers = []

for cluster_id in np.unique(clusters):
    cluster_mask = (clusters == cluster_id)
    cluster_center = np.mean(data[cluster_mask], axis=0)
    cluster_centers.append(cluster_center)

cluster_centers = np.array(cluster_centers)


# Визуализация центров кластеров
plt.figure(figsize=(8, 6))
plt.scatter(df['X'], df['Y'], c=clusters, cmap='nipy_spectral', s=50)
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], c='red', marker='x', s=100, label='Centers')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('График кластеризации и центры кластеров')
plt.show()

#single + euclid 0.81
#single + seuclid 0.835
#single + minkowski 0.791
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist

#  Минк расст с p = 4
def minkowski_distance(x, y):
    return np.sum(np.abs(x - y) ** 4) ** (1/4)


with open('/content/2023_2.txt', 'r') as file:
    lines = file.readlines()

data = [line.strip().split() for line in lines]
df = pd.DataFrame(data, columns=['X', 'Y'])
df[['X', 'Y']] = df[['X', 'Y']].apply(pd.to_numeric)

plt.scatter(df['X'], df['Y'])
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Диаграмма рассеяния данных')
plt.show()

# алгоритм кластеризации методом k-средних
def k_means_clustering(data, k):
    kmeans = KMeans(n_clusters=k, init='k-means++', n_init=10, max_iter=300, random_state=0)
    kmeans.inertia_ = None  # Очищаем инерцию, чтобы использовать свою функцию расстояния
    kmeans.distance = minkowski_distance  # Используем собственную функцию расстояния

    kmeans.fit(data)
    return kmeans

#кластерный анализ исходных данных
distortions = []  # список для хранения искажений
K = range(1, 11)

for k in K:
    kmeans = k_means_clustering(df[['X', 'Y']].to_numpy(), k)
    distortions.append(sum(np.min(cdist(df[['X', 'Y']].to_numpy(), kmeans.cluster_centers_), axis=1)) / df.shape[0])


plt.plot(K, distortions, marker='o')
plt.xlabel('Количество кластеров')
plt.ylabel('Искажение')
plt.title('Метод локтя')
plt.show()


optimal_k = 4  #  оптимальное число кластеров
print(f"Оптимальное число кластеров k: {optimal_k}")

# кластеризация с оптимальным k
kmeans = k_means_clustering(df[['X', 'Y']].to_numpy(), optimal_k)
df['Cluster'] = kmeans.labels_

cluster_centers = kmeans.cluster_centers_


plt.scatter(df['X'], df['Y'], c=df['Cluster'], cmap='viridis')
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], c='red', marker='x', s=100, label='Centers')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('График кластеризации методом k-средних с Минковским расстоянием и p=4')
plt.show()
```
